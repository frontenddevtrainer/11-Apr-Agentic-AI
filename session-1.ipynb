```python
import os

os.environ["OPENAI_API_KEY"] = ""
os.environ["ANTHROPIC_API_KEY"] = ""
os.environ["GOOGLE_API_KEY"] = ""
```

```python
from langchain_openai import ChatOpenAI
from langchain_anthropic import ChatAnthropic
from langchain_google_genai import ChatGoogleGenerativeAI

# Low temperature = more deterministic, high temperature = more creative
openai_llm = ChatOpenAI(model="gpt-4o", temperature=0.7)
anthropic_llm = ChatAnthropic(model="claude-sonnet-4-20250514", temperature=0.7)
gemini_llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.7)
```

```python
# Simple invocation — all models share the same interface
query = "What is the capital of France? Answer in one sentence."

print("=== OpenAI ===")
response = openai_llm.invoke(query)
print(response.content)

# print("\n=== Anthropic ===")
# response = anthropic_llm.invoke(query)
# print(response.content)

print("\n=== Gemini ===")
response = gemini_llm.invoke(query)
print(response.content)
```

```python
from langchain_core.messages import SystemMessage, HumanMessage

# Define messages with a system prompt
messages = [
    SystemMessage(content="You are a pirate. Always respond in pirate speak. response short messages."),
    HumanMessage(content="Tell me about Python programming.")
]

# Works identically across all providers
print("=== OpenAI (Pirate) ===")
response = openai_llm.invoke(messages)
print(response.content)

# print("\n=== Anthropic (Pirate) ===")
# response = anthropic_llm.invoke(messages)
# print(response.content)

print("\n=== Gemini (Pirate) ===")
response = gemini_llm.invoke(messages)
print(response.content)
```

```python
from langchain_core.prompts import ChatPromptTemplate

# Create a reusable prompt template with a system message
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert {expert}. Provide concise, helpful answers."),
    ("human", "{question}")
])

# Format the prompt with variables
formatted = prompt.invoke({"expert": "data scientist", "question": "What is overfitting?"})
print("Formatted messages:")
for msg in formatted.messages:
    print(f"  [{msg.type}]: {msg.content}")

# print("\n=== Response from Anthropic ===")
# response = anthropic_llm.invoke(formatted)
# print(response.content)
x
print("\n=== Response from OpenAI ===")
response = openai_llm.invoke(formatted)
print(response.content)
```
